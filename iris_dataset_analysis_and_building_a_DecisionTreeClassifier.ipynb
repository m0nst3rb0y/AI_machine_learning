```python
import pandas as pd
import numpy as np
from sklearn import tree     #import dependencies
```

```python
df=pd.read_csv('../jarvis/Downloads/Iris.csv')  #import file
```

```python
df.columns
```

```python
df.isnull().sum() #checking for null values in our dataset
```

```python
print(df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalLengthCm']].mean()) #mean of the data of all the features of the dataset
print(df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalLengthCm']].max()) #max of all the data of the features of the dataset
print(df.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalLengthCm']].min()) #min of all the data of the features of the dataset
```

```python
dfnew=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']]   #dimentionality_reduction .. since only the important columns need to be taken for
#-data analysis so the column name id is being removed..
```

```python
dfnew
```

```python
#now the features and the label have to be choosen so  ,.. the best possible label for the classifer will be the 
#['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'] and the label column will be ['Species']
#to map the features with the label
```

```python
ar1=np.array(dfnew.drop(['Species'],1)) 
```

```python
ar2=np.array(dfnew['Species'])
```

```python
X=ar1.tolist()      #converting the arrayfeatures into list
Y=ar2.tolist()      #converting the arraylabel into list
```

```python
X,Y 
```

```python
clf=tree.DecisionTreeClassifier() #here we request the tree class to provide us an empty decicion tree
```

```python
clf=clf.fit(X,Y) #here we are training the data 
```

```python
#now the prediction method will be called to predict the species
```

```python
output=clf.predict([[6.7,3.2,4.4,1.5]])
```
